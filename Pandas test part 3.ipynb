```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\RISHU YADAV\Downloads\framingham.csv')
```

```python
df
```

```python
# Question (1):Go to framingum.csv and Take those People who are Male and  his education level is 3 and Heart Rate is grater than 65.

df.loc[(df['male']==1) & (df['education']==3)&(df['heartRate']>65),['male','education','heartRate']]
```

```python
# Question (1):Repeat the Question 1 for Female

df.loc[(df['male']==0) & (df['education']==3)&(df['heartRate']>65),['male','education','heartRate']]
```

```python
# Question (3) Write group by for making male as index and citrate per day in values for mean analysis

df.groupby(['male','cigsPerDay']).agg({'heartRate':['mean']})
```

```python
# Question (4) Repeat the Question 3 for Pivot

table = pd.pivot_table(df, values='heartRate', index=['male','cigsPerDay'], aggfunc=np.mean) 
```
