```python
x=[12321,232,4325,34543,56787,456,545,345,34643,1298921]
for i in x:
    rev=0
    y=i
    while(i>0):
        k=i%10
        rev=rev*10+k
        i=i//10
    if(rev==y):
        print("palindrome",y)
    else:
        print("Not palindrome",y)
    
```
